```python
taxis = pd.read_parquet('https://github.com/j-rossi-nl/teaching-data/raw/main/2022_IDM/nyc_taxi_small.parq').dropna()
taxis['count'] = 1

location = pd.read_csv('https://s3.amazonaws.com/nyc-tlc/misc/taxi+_zone_lookup.csv')

answers.__safe__taxis = (taxis, '33e26a0ca934485c1ada7260eb7ac75f5c5973827c14474e879cb7c2d887e812')
answers.__safe__location = (location, 'f74c986fed3c6f51f55b61fd7a50a5dfe3e3ed61cb0e3063398737478e1d1c5e')
taxis
```

```python
# 1.feladat
if 'level_0' in taxis:
  taxis.drop(columns=['level_0'], inplace=True)
taxis.reset_index(inplace=True)
tip_pct=[]
for i in range(len(taxis)):
  tip=taxis._get_value(i,'tip_amount')
  total=taxis._get_value(i,'total_amount')
  pct=(tip/total)*100
  tip_pct.append(pct)

taxis['tip_pct']=tip_pct
taxis
```

```python
# 2.Feladat
good_tips=taxis.loc[taxis['total_amount']>0]
good_tips
```

```python
# 3. Feladat
from collections import Counter
tmp3=Counter(taxis['PULocationID'])
minlim=50
data_zones=[]
for key,count in tmp3.items():
  if count>=minlim:
    data_zones.append(key)

data_zones
```

```python
# 4. Feladat
new=taxis['PULocationID'].isin(data_zones)
dct={}
df=taxis[new].groupby('PULocationID')
for item in data_zones:
  avg=sum(df.get_group(item)['trip_distance'])/len(df.get_group(item))
  dct[item]=avg

tmp4=[]
for value in dct.values():
  tmp4.append(value)

tmp4.sort()
tmp4=tmp4[:10]
shortest_triptmp=[]
for key,value in dct.items():
  for j in range(len(tmp4)):
    if value==tmp4[j]:
      shortest_triptmp.append(key)

shortest_trip=pd.Series(shortest_triptmp)
shortest_trip
```

```python
# 5. Feladat
tmp5=[]
for i in range(len(taxis)):
  pickup=taxis._get_value(i,'PULocationID')
  dropoff=taxis._get_value(i,'DOLocationID')
  tmp5.append(f'{pickup} - {dropoff}')

howmany=Counter(tmp5)
top_pairstmp=[]
for item in howmany.most_common(20):
  top_pairstmp.append(item[0])

top_pairs=pd.Series(top_pairstmp)
top_pairs
```

```python
# 6.Feladat
if 'level_0' in good_tips:
  good_tips.drop(columns=['level_0'], inplace=True)
good_tips.reset_index(inplace=True)
tempo6=[]
tmp6={}
for i in range(len(good_tips)):
  tip=good_tips._get_value(i,'tip_amount')
  pickup=good_tips._get_value(i,'PULocationID')
  dropoff=good_tips._get_value(i,'DOLocationID')
  tempo6.append(f'{pickup} - {dropoff}')
  if f'{pickup} - {dropoff}' in tmp6:
    tmp6[f'{pickup} - {dropoff}']+=tip
  else:
    tmp6[f'{pickup} - {dropoff}']=tip

trip_numbers=Counter(tempo6)
for key,value in tmp6.items():
  times=trip_numbers[key]
  avg_tip=value/times
  tmp6[key]=avg_tip

tempo6=[]
for value in tmp6.values():
  tempo6.append(value)

tempo6.sort()
toptipslist=tempo6[-10:]
top_tipstmp=[]
for key,value in tmp6.items():
  for j in range(len(toptipslist)):
    if value==toptipslist[j]:
      if key not in top_tipstmp:
        top_tipstmp.append(key)

top_tips=pd.Series(top_tipstmp)
top_tips
```

```python
# 7. Feladat
zone_pair=[]
for i in range(len(taxis)):
  pickup=taxis._get_value(i,'PULocationID')
  dropoff=taxis._get_value(i,'DOLocationID')
  zone_pair.append(f'{pickup}-{dropoff}')

taxis['zone_pair']=zone_pair
good_tips=taxis.loc[taxis['total_amount']>0]
if 'level_0' in good_tips:
  good_tips.drop(columns=['level_0'], inplace=True)
good_tips.reset_index(inplace=True)
minlim=50
tmp7=Counter(good_tips['zone_pair'])
zones_above=[]
for key,count in tmp7.items():
  if count>=minlim:
    zones_above.append(key)

fitting=taxis['zone_pair'].isin(zones_above)
temp_df=taxis[fitting]
if 'level_0' in temp_df:
  temp_df.drop(columns=['level_0'], inplace=True)
temp_df.reset_index(inplace=True)
tempo7=[]
temp7={}
for i in range(len(temp_df)):
  tip=temp_df._get_value(i,'tip_amount')
  pickup=temp_df._get_value(i,'PULocationID')
  dropoff=temp_df._get_value(i,'DOLocationID')
  tempo7.append(f'{pickup} - {dropoff}')
  if f'{pickup} - {dropoff}' in temp7:
    temp7[f'{pickup} - {dropoff}']+=tip
  else:
    temp7[f'{pickup} - {dropoff}']=tip

trip_numbers=Counter(tempo7)
for key,value in temp7.items():
  times=trip_numbers[key]
  avg_tip=value/times
  temp7[key]=avg_tip

tempo7=[]
for value in temp7.values():
  tempo7.append(value)

tempo7.sort()
toptipslist=tempo7[-10:]
reliable_tipstmp=[]
for key,value in temp7.items():
  for j in range(len(toptipslist)):
    if value==toptipslist[j]:
      if key not in reliable_tipstmp:
        reliable_tipstmp.append(key)

reliable_tips=pd.Series(reliable_tipstmp)
reliable_tips
```

```python
# 8. Feladat
manhattan=[]
for i in range(len(location)):
  if location._get_value(i,'Borough')=='Manhattan':
    id=location._get_value(i,'LocationID')
    manhattan.append(id)

manhattan
```

```python
# 9. Feladat
filter=taxis['PULocationID'].isin(manhattan)
manhattan_start=taxis[filter]
start_day=manhattan_start['tpep_pickup_datetime'].dt.day_name().tolist()
start_hour=manhattan_start['tpep_pickup_datetime'].dt.hour.tolist()
manhattan_start['start_day']=start_day
manhattan_start['start_hour']=start_hour
if 'level_0' in manhattan_start:
  manhattan_start.drop(columns=['level_0'], inplace=True)
manhattan_start.reset_index(inplace=True)
pickups=[]
for i in range(len(manhattan_start)):
  zone=manhattan_start._get_value(i,'PULocationID')
  day=manhattan_start._get_value(i,'start_day')
  hour=manhattan_start._get_value(i,'start_hour')
  pickups.append(f'{zone} - {day}/{hour}')

counts=Counter(pickups)
tmp9=[]
for item in counts.most_common(20):
  tmp9.append(item[0])

top_busy=pd.Series(tmp9)
top_busy
```

```python
# 10.feladat
for i in range(len(location)):
  if location._get_value(i,'Zone')=='JFK Airport':
    jfk_loc_id=location._get_value(i,'LocationID')
    break

jfk_dest_filter=manhattan_start['DOLocationID'].isin([jfk_loc_id])
manh_to_jfk=manhattan_start[jfk_dest_filter]
if 'level_0' in manh_to_jfk:
  manh_to_jfk.drop(columns=['level_0'], inplace=True)
manh_to_jfk.reset_index(inplace=True)
pickups_new=[]
for i in range(len(manh_to_jfk)):
  zone=manh_to_jfk._get_value(i,'PULocationID')
  day=manh_to_jfk._get_value(i,'start_day')
  hour=manh_to_jfk._get_value(i,'start_hour')
  pickups_new.append(f'{zone} - {day}/{hour}')

counts_new=Counter(pickups_new)
tmp10=[]
for item in counts_new.most_common(5):
  tmp10.append(item[0])

top_manhattan_jfk=pd.Series(tmp10)
top_manhattan_jfk
```
